# Theory Ques.

Q1 What is hypothesis testing in statistics

Hypothesis Testing is a fundamental statistical method used to make decisions or inferences about a population based on a sample of data. It helps determine whether there is enough evidence in a sample to support or reject a specific assumption (called a hypothesis) about a population parameter.

Q2 What is the null hypothesis, and how does it differ from the alternative hypothesis

Null Hypothesis (H‚ÇÄ):

It is the default assumption that there is no effect, no difference, or no relationship in the population.

Example: "The mean height of students is 170 cm." (H‚ÇÄ: Œº = 170)

Alternative Hypothesis (H‚ÇÅ or Ha):

It contradicts the null and proposes that there is an effect, difference, or relationship.

Example: "The mean height is not 170 cm." (H‚ÇÅ: Œº ‚â† 170)

Difference:

H‚ÇÄ assumes the status quo; H‚ÇÅ challenges it.

Hypothesis testing determines whether to reject H‚ÇÄ in favor of H‚ÇÅ.



Q3 What is the significance level in hypothesis testing, and why is it important?

Significance Level (Œ±): The probability of rejecting the null hypothesis when it is actually true (Type I error).

Common values: 0.05, 0.01, or 0.10

Importance:

Sets the threshold for deciding whether results are statistically significant.

Helps control the risk of false positives.

Q4 What does a P-value represent in hypothesis testing ?

The P-value is the probability of obtaining a test statistic as extreme as the one observed, assuming the null hypothesis is true.

It quantifies how much evidence we have against H‚ÇÄ.

Q5 How do you interpret the P-value in hypothesis testing ?

If P ‚â§ Œ± ‚Üí Reject the null hypothesis ‚Üí Strong evidence against H‚ÇÄ

If P > Œ± ‚Üí Fail to reject H‚ÇÄ ‚Üí Not enough evidence to reject H‚ÇÄ

Example:

P-value = 0.03, Œ± = 0.05 ‚Üí Reject H‚ÇÄ (statistically significant)

Q6 What are Type I and Type 2 errors in hypothesis testing ?

Type I Error (False Positive):

Rejecting H‚ÇÄ when it is actually true

Probability = Œ±

Type II Error (False Negative):

Failing to reject H‚ÇÄ when it is actually false

Probability = Œ≤

Q7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?

One-tailed test: Tests for deviation in one direction (e.g., greater than or less than)

Example: H‚ÇÅ: Œº > 100

Two-tailed test: Tests for deviation in both directions (e.g., not equal)

Example: H‚ÇÅ: Œº ‚â† 100

Q8 What is the Z-test, and when is it used in hypothesis testing ?

A Z-test is used to determine whether there is a significant difference between sample and population means when:

The population standard deviation is known

The sample size is large (n > 30)

Example: Testing whether average income of a city is ‚Çπ50,000.

Q1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy import stats

# Sample data (example)
sample_data = [52, 49, 50, 51, 53, 48, 54, 47, 50, 49]

# Step 1: Define known values
population_mean = 50     # Known population mean
population_std = 2       # Known population standard deviation (œÉ)
sample_size = len(sample_data)

# Step 2: Calculate sample mean
sample_mean = np.mean(sample_data)

# Step 3: Compute Z-score using formula:
# Z = (sample_mean - population_mean) / (population_std / sqrt(n))
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Calculate P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 5: Print results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_score, 4))
print("P-value:", round(p_value, 4))

# Step 6: Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Result: Reject the null hypothesis (Significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (No significant difference).")
```

Q4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Sample data (you can customize this)
sample_data = [102, 98, 105, 101, 100, 99, 103, 97, 106, 100]

# Step 2: Known population parameters
population_mean = 100
population_std = 3  # Known œÉ

# Step 3: Calculate statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# Step 4: Z critical value for 95% confidence (Œ± = 0.05)
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

# Step 5: Print Results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_score, 4))
print("P-value:", round(p_value, 4))
if abs(z_score) > z_critical:
    print("Conclusion: Reject the null hypothesis ‚Äî significant difference.")
else:
    print("Conclusion: Fail to reject the null hypothesis ‚Äî no significant difference.")

# Step 6: Visualization of the decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (Right)')

# Mark Z-score
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')

# Mark critical values
plt.axvline(-z_critical, color='black', linestyle=':', label=f'-Z-critical = {-z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':', label=f'Z-critical = {z_critical:.2f}')

plt.title("Two-Tailed Z-Test: Decision Region")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

Q5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu_null=100, mu_alt=105, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a right-tailed Z-test.

    Parameters:
    - mu_null: Mean under the null hypothesis (H‚ÇÄ)
    - mu_alt: Mean under the alternative hypothesis (H‚ÇÅ)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (default 0.05 for 95% confidence)
    """

    # Step 1: Calculate standard error
    se = sigma / np.sqrt(n)

    # Step 2: Z critical value and critical sample mean for right-tailed test
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu_null + z_critical * se

    # Step 3: Calculate Type II error (Œ≤)
    z_beta = (x_critical - mu_alt) / se
    beta = norm.cdf(z_beta)

    # Step 4: Calculate power of the test
    power = 1 - beta

    # Step 5: Plot null and alternative distributions
    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)
    y_null = norm.pdf(x, mu_null, se)
    y_alt = norm.pdf(x, mu_alt, se)

    plt.figure(figsize=(12, 6))
    plt.plot(x, y_null, label='Null Hypothesis (H‚ÇÄ)', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H‚ÇÅ)', color='green')

    # Step 6: Shade rejection region (Type I error)
    plt.fill_between(x, y_null, where=(x > x_critical), color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Step 7: Shade Type II error region
    plt.fill_between(x, y_alt, where=(x < x_critical), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    # Step 8: Critical value line
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value = {x_critical:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing', fontsize=14)
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Step 9: Print interpretation
    print(f"‚û°Ô∏è  Critical Value (Sample Mean): {x_critical:.2f}")
    print(f"‚û°Ô∏è  Z-critical value: {z_critical:.4f}")
    print(f"‚ö†Ô∏è  Type I Error (Œ±): {alpha}")
    print(f"‚ö†Ô∏è  Type II Error (Œ≤): {beta:.4f}")
    print(f"‚úÖ  Power of the Test (1 - Œ≤): {power:.4f}")

# Example usage
visualize_type1_type2(mu_null=100, mu_alt=105, sigma=10, n=30, alpha=0.05)
```

Q6 Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
from scipy import stats

# Step 1: Define two independent samples (e.g., test scores of two classes)
group1 = [85, 86, 88, 75, 78, 94, 98, 79, 81, 87]
group2 = [80, 82, 84, 76, 77, 88, 85, 83, 79, 78]

# Step 2: Perform independent two-sample t-test
# Use equal_var=False for Welch's t-test if variances are unequal
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)

# Step 3: Output results
print("Group 1 Mean:", np.mean(group1))
print("Group 2 Mean:", np.mean(group2))
print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

# Step 4: Interpret results (two-tailed test)
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis ‚Äî significant difference between the groups.")
else:
    print("Conclusion: Fail to reject the null hypothesis ‚Äî no significant difference between the groups.")
```

Q9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample.
    
    Parameters:
    - data (list or array): The sample data
    - confidence (float): Confidence level (default = 0.95 for 95%)
    
    Returns:
    - (lower_bound, upper_bound): The confidence interval
    """

    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard Error of the Mean
    margin = stats.t.ppf((1 + confidence) / 2.0, n - 1) * sem

    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"Sample Size: {n}")
    print(f"Confidence Level: {confidence * 100:.1f}%")
    print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return (lower, upper)

# ‚úÖ Example usage
sample_data = [72, 74, 78, 76, 75, 80, 79, 82, 77, 74]
confidence_interval(sample_data, confidence=0.95)
```

Q10  Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.
    
    Parameters:
    - data: list or numpy array of sample values
    - confidence: confidence level (default = 0.95 for 95%)
    
    Returns:
    - Margin of Error (MoE)
    """

    data = np.array(data)
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation
    standard_error = sample_std / np.sqrt(n)

    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    # Margin of Error
    moe = t_critical * standard_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}")
    print(f"Sample Size: {n}")
    print(f"Confidence Level: {confidence * 100:.1f}%")
    print(f"Margin of Error: ¬±{moe:.2f}")

    return moe

# ‚úÖ Example usage
sample_data = [72, 74, 78, 76, 75, 80, 79, 82, 77, 74]
calculate_margin_of_error(sample_data, confidence=0.95)
```

Q11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):
    """
    Apply Bayes' Theorem to compute the posterior probability.

    Parameters:
    - prior_A: P(A), prior probability of hypothesis A (e.g., disease)
    - likelihood_B_given_A: P(B|A), likelihood of evidence given A (e.g., test is positive if person has disease)
    - likelihood_B_given_not_A: P(B|¬¨A), likelihood of evidence given not A (false positive rate)

    Returns:
    - posterior: P(A|B), updated belief in A given evidence B
    """
    prior_not_A = 1 - prior_A
    # Marginal probability of evidence
    evidence = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    
    # Posterior probability
    posterior = (likelihood_B_given_A * prior_A) / evidence
    
    print(f"Posterior Probability (P(A|B)): {posterior:.4f}")
    return posterior

# Example usage
# Disease example
prior_disease = 0.01
likelihood_positive_given_disease = 0.99
likelihood_positive_given_healthy = 0.05

bayes_theorem(prior_disease, likelihood_positive_given_disease, likelihood_positive_given_healthy)
```

Q12 Perform a Chi-square test for independence between two categorical variables in Python

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (observed frequencies)
# Rows = Gender (Male, Female), Columns = Product Preference (A, B, C)
data = [[30, 10, 20],    # Male
        [25, 15, 30]]    # Female

# Convert to DataFrame for clarity
df = pd.DataFrame(data, columns=["Product A", "Product B", "Product C"], index=["Male", "Female"])

print("Contingency Table:")
print(df)

# Step 2: Perform Chi-Square Test of Independence
chi2, p, dof, expected = chi2_contingency(df)

# Step 3: Display results
print("\nChi-square Statistic:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))

# Show expected frequencies
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=df.columns, index=df.index))

# Step 4: Interpret the result
alpha = 0.05
if p < alpha:
    print("\nConclusion: Reject the null hypothesis ‚Äî variables are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis ‚Äî variables are independent.")
```

Q16 Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two samples.
    
    Parameters:
    - sample1: List or array of sample 1
    - sample2: List or array of sample 2
    - alpha: Significance level (default = 0.05)
    
    Returns:
    - Prints F-statistic, p-value, and conclusion
    """
    # Step 1: Sample sizes and variances
    n1, n2 = len(sample1), len(sample2)
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)

    # Step 2: Compute F-statistic (larger variance on top)
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Step 3: Calculate p-value (two-tailed)
    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    # Step 4: Results
    print("Sample 1 Variance:", round(var1, 4))
    print("Sample 2 Variance:", round(var2, 4))
    print("F-statistic:", round(f_stat, 4))
    print("Degrees of Freedom:", (dfn, dfd))
    print("P-value:", round(p_value, 4))

    # Step 5: Conclusion
    if p_value < alpha:
        print("Conclusion: Reject H‚ÇÄ ‚Äî Variances are significantly different.")
    else:
        print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference in variances.")

# ‚úÖ Example usage
np.random.seed(42)
sample_1 = np.random.normal(loc=50, scale=5, size=30)
sample_2 = np.random.normal(loc=52, scale=10, size=30)

f_test(sample_1, sample_2, alpha=0.05)
```

Q19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups, group_names=None, alpha=0.05):
    """
    Check ANOVA assumptions: normality (Shapiro-Wilk) and equal variances (Levene's test).
    
    Parameters:
    - groups: list of lists or numpy arrays (e.g., [group1, group2, group3])
    - group_names: optional list of group names (e.g., ["Group A", "Group B"])
    - alpha: significance level for tests (default is 0.05)
    """

    if group_names is None:
        group_names = [f"Group {i+1}" for i in range(len(groups))]

    print("\nüîç Checking ANOVA Assumptions\n" + "-"*35)

    # ‚úÖ 1. Normality check (Shapiro-Wilk Test)
    print("\n‚úÖ Normality Check (Shapiro-Wilk Test):")
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        status = "‚úîÔ∏è Normally distributed" if p > alpha else "‚ùå Not normally distributed"
        print(f"{group_names[i]}: p-value = {p:.4f} ‚Üí {status}")

    # ‚úÖ 2. Equal variance check (Levene‚Äôs Test)
    print("\n‚úÖ Homogeneity of Variance (Levene‚Äôs Test):")
    stat, p = levene(*groups)
    status = "‚úîÔ∏è Variances are equal" if p > alpha else "‚ùå Variances are significantly different"
    print(f"Levene‚Äôs test p-value = {p:.4f} ‚Üí {status}")

    # ‚úÖ 3. Independence (Not testable here)
    print("\n‚ÑπÔ∏è Independence Assumption:")
    print("Ensure that observations are independently collected (e.g., random sampling, no overlap).")

# ----------------------------
# ‚úÖ Example Usage
# ----------------------------

group_A = [85, 90, 88, 86, 87]
group_B = [78, 75, 80, 77, 76]
group_C = [92, 94, 89, 91, 95]

check_anova_assumptions([group_A, group_B, group_C], group_names=["Group A", "Group B", "Group C"])
```

Q21  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(df1, df2, alpha=0.05):
    """
    Visualize the F-distribution with critical region.
    
    Parameters:
    - df1: degrees of freedom (numerator)
    - df2: degrees of freedom (denominator)
    - alpha: significance level (default = 0.05)
    """
    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, df1, df2)

    # Critical value (right-tail)
    f_critical = f.ppf(1 - alpha, df1, df2)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')

    # Shade critical region
    x_crit = np.linspace(f_critical, 5, 300)
    plt.fill_between(x_crit, f.pdf(x_crit, df1, df2), color='red', alpha=0.3, label='Rejection region')

    plt.axvline(f_critical, color='red', linestyle='--', label=f'Critical Value = {f_critical:.2f}')
    plt.title("F-Distribution with Rejection Region")
    plt.xlabel("F-Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print(f"F Critical Value at Œ± = {alpha}: {f_critical:.4f}")

# ‚úÖ Example usage
plot_f_distribution(df1=5, df2=20, alpha=0.05)
```

Q24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=25)  # True std dev = 5
sample_variance = np.var(sample, ddof=1)
n = len(sample)

# Step 2: Hypothesized population variance
sigma_0 = 5  # Hypothesized population standard deviation
sigma_0_squared = sigma_0**2

# Step 3: Calculate Chi-square statistic
chi_square_stat = (n - 1) * sample_variance / sigma_0_squared
df = n - 1  # degrees of freedom

# Step 4: Compute critical values for two-tailed test
alpha = 0.05
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Step 5: Print results
print("Sample Variance:", round(sample_variance, 4))
print("Hypothesized Variance:", sigma_0_squared)
print("Chi-square Statistic:", round(chi_square_stat, 4))
print(f"Chi-square Critical Values: Lower = {round(chi2_lower, 4)}, Upper = {round(chi2_upper, 4)}")

# Step 6: Decision
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Conclusion: ‚ùå Reject H‚ÇÄ ‚Äî Population variance is significantly different from hypothesized.")
else:
    print("Conclusion: ‚úÖ Fail to reject H‚ÇÄ ‚Äî No significant difference in population variance.")
```

Q25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(success_a, size_a, success_b, size_b, alpha=0.05):
    """
    Performs a two-proportion Z-test and prints the result.
    
    Parameters:
    - success_a, size_a: Successes and total trials in group A
    - success_b, size_b: Successes and total trials in group B
    - alpha: Significance level (default = 0.05)
    """
    # Successes and observations
    counts = np.array([success_a, success_b])
    nobs = np.array([size_a, size_b])

    # Perform Z-test
    z_stat, p_value = proportions_ztest(count=counts, nobs=nobs)

    print("‚úÖ Z-Test for Two Proportions")
    print(f"Group A: {success_a}/{size_a} ‚Üí Proportion = {success_a/size_a:.4f}")
    print(f"Group B: {success_b}/{size_b} ‚Üí Proportion = {success_b/size_b:.4f}")
    print(f"\nZ-statistic = {z_stat:.4f}")
    print(f"P-value     = {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(f"Conclusion: ‚ùå Reject H‚ÇÄ ‚Äî Significant difference in proportions at Œ± = {alpha}")
    else:
        print(f"Conclusion: ‚úÖ Fail to reject H‚ÇÄ ‚Äî No significant difference in proportions at Œ± = {alpha}")

# --------------------------------------------
# ‚úÖ Example usage
# Group A: 60 successes out of 200
# Group B: 72 successes out of 250
# --------------------------------------------
z_test_proportions(success_a=60, size_a=200, success_b=72, size_b=250, alpha=0.05)
```

Q26  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def f_test_variances(data1, data2, alpha=0.05):
    # Step 1: Calculate sample variances
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)

    # Step 2: Determine which is larger to keep F > 1
    if var1 > var2:
        F = var1 / var2
        dfn = len(data1) - 1
        dfd = len(data2) - 1
    else:
        F = var2 / var1
        dfn = len(data2) - 1
        dfd = len(data1) - 1

    # Step 3: Critical values for two-tailed test
    f_critical_low = f.ppf(alpha / 2, dfn, dfd)
    f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

    print("‚úÖ F-Test for Equal Variances")
    print(f"Variance 1 = {var1:.4f}")
    print(f"Variance 2 = {var2:.4f}")
    print(f"F-statistic = {F:.4f}")
    print(f"Degrees of freedom: df1 = {dfn}, df2 = {dfd}")
    print(f"Critical F-values: Lower = {f_critical_low:.4f}, Upper = {f_critical_high:.4f}")

    if F < f_critical_low or F > f_critical_high:
        print(f"‚ùå Reject H‚ÇÄ ‚Üí Variances are significantly different at Œ± = {alpha}")
    else:
        print(f"‚úÖ Fail to reject H‚ÇÄ ‚Üí No significant difference in variances at Œ± = {alpha}")

    # Step 4: Plot F-distribution and critical regions
    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, dfn, dfd)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='F-distribution', color='blue')
    plt.axvline(F, color='green', linestyle='--', label=f'F-stat = {F:.2f}')
    plt.axvline(f_critical_low, color='red', linestyle='--', label='Critical Lower')
    plt.axvline(f_critical_high, color='red', linestyle='--', label='Critical Upper')

    # Shade critical regions
    plt.fill_between(x, 0, y, where=(x < f_critical_low) | (x > f_critical_high), color='red', alpha=0.3)
    
    plt.title("F-Test for Comparing Two Variances")
    plt.xlabel("F value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ------------------------------------
# ‚úÖ Example usage
# Simulate data from normal distributions with different variances
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)  # Std dev = 10
group2 = np.random.normal(loc=100, scale=15, size=30)  # Std dev = 15

f_test_variances(group1, group2, alpha=0.05)

```
